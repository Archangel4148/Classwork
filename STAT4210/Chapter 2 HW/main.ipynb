```python
college = college3
```

```python
college.describe()
```

```python
pd.plotting.scatter_matrix(college[["Top10perc", "Apps", "Enroll"]])
```

```python
college["Elite"] = pd.cut(college["Top10perc"]/100, [0,0.5,1], labels=["No", "Yes"]) 
college["Elite"].value_counts() 
college.boxplot("Outstate", by="Elite")
```
